```python
import pandas as pd
```

```python
df = pd.read_csv("Amazon-Dataset.csv", encoding="ISO-8859-1")
df.info(), df.head()
```

```python
df["Postal Code"].fillna(0, inplace=True)
df.info(), df.head()
```

```python
print(df[["Order ID", "Order Date"]].dtypes)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"],format= '%d/%m/%Y')
df["Ship Date"] = pd.to_datetime(df["Ship Date"],format= '%d/%m/%Y')
df.info(), df.head()
```

```python
df['Sales'] = df['Sales'].replace('[\$,]', '', regex=True).astype(float)
df['Profit'] = df['Profit'].replace('[\$,]', '', regex=True).astype(float)
df.info(), df.head()
```

```python
summary_statistics  = df[["Sales", "Profit", "Shipping Cost"]].describe()
```

```python
summary_statistics
```

```python
segment_distribution = df["Segment"].value_counts()
market_distribution = df["Market"].value_counts()
segment_distribution
```

```python
market_distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (12,5))
sns.barplot(x= segment_distribution.index, y = segment_distribution.values, palette = "viridis")
plt.title  ("Distribution of Orders across Customer Segments")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Orders")
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x= market_distribution.index, y = market_distribution.values, palette = "viridis")
plt.title  ("Distribution of Orders across Market Segments")
plt.xlabel("Market Segment")
plt.ylabel("Number of Orders")
plt.show()
```

```python
categoryProfit = df.groupby(["Category", "Sub-Category"])["Profit"].sum().reset_index()
highestProfitCategory = categoryProfit.loc[categoryProfit['Profit'].idxmax()]
highestProfitCategory
```

```python
productProfit = df.groupby('Product Name')['Profit'].sum().reset_index()
top5Products = productProfit.sort_values(by= 'Profit', ascending=False).head(5)
top5Products
```

```python
countryRegionSales = df.groupby(['Country', 'Region'])['Sales'].sum().reset_index()
countryRegionSales
```

```python
highestSalesCountryRegion = countryRegionSales.loc[countryRegionSales['Sales'].idxmax()]
highestSalesCountryRegion
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Shipping Cost', y='Profit', hue='Region', palette='viridis', alpha=0.7)
plt.title('Relationship between Shipping Cost and Profit by Region')
plt.xlabel('Shipping Cost')
plt.ylabel('Profit')
plt.legend(title='Region',bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```

```python
segmentRevenue = df.groupby('Segment')['Sales'].sum().reset_index()
segmentRevenue
```

```python
highestRevenue = segmentRevenue.loc[segmentRevenue['Sales'].idxmax()]
print("Segment contributes the most to revenue is: ")
highestRevenue
```

```python
segmentStats = df.groupby('Segment').agg(avgDiscount=('Discount','mean'), avgProfit=('Profit','mean')).reset_index()
segmentStats
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Segment', palette='Set2', alpha=0.7)
plt.title('Impact of Discount on Profit by Customer Segment')
plt.xlabel('Discount %')
plt.ylabel('Profit')
plt.legend(title='Segment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month']= df['Order Date'].dt.month

monthlySales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
monthlySales
```

```python
monthlySales['YearMonth'] = monthlySales['Year'].astype(str) + '-' + monthlySales['Month'].astype(str)
plt.figure(figsize=(12,6))
sns.lineplot(data=monthlySales, x='YearMonth', y='Sales', marker = 'o')
plt.title('Monthly Sales Trends')
plt.xlabel('Year Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()
yearly_comparison = yearly_sales[yearly_sales['Year'].isin([2012, 2015])]
print("\nYearly Sales Comparison (2012 vs. 2015):")
print(yearly_comparison)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=yearly_sales, x='Year', y='Sales', palette='viridis')
plt.title('Yearly Sales Performance')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

# Three Actionable Strategies to Improve Profitability

## 1. Optimize Discounts and Pricing Strategy  
Your scatterplot on **discount vs. profit** suggests that **higher discounts may be negatively impacting profitability**.  

### Actionable Steps:  
- Identify the **optimal discount range** that maximizes profit without eroding margins.  
- Implement **targeted promotions** instead of blanket discounts, focusing on **high-margin products**.  
- Experiment with **dynamic pricing** to adjust prices based on demand and seasonality.  

## 2. Improve Shipping Cost Efficiency  
The scatterplot on **shipping cost vs. profit** suggests a correlation where **high shipping costs might be cutting into profits**.  

### Actionable Steps:  
- **Negotiate better shipping rates** with carriers or explore alternative logistics providers.  
- Introduce **regional warehouses** in high-sales areas to **reduce shipping distances**.  
- Offer customers **incentives for bulk purchases** or slower, cost-efficient shipping options.  

## 3. Focus on High-Profit Products and Regions  
Your analysis identified the **most profitable categories, sub-categories, and products**.  

### Actionable Steps:  
- Increase **marketing efforts for top-performing products** while reconsidering or optimizing underperforming ones.  
- Expand sales efforts in **high-sales regions** while investigating and addressing challenges in low-sales areas.  
- **Personalize promotions** based on **customer segments** that contribute the most revenue.
